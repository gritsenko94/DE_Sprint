```python
# 1 Парсер hh
import requests as req
from bs4 import BeautifulSoup
import os
import unidecode
import json


# получаем все урлы с вакансиями
vacancies = []
url = f'https://hh.ru/search/vacancy?no_magic=true&L_save_area=true&text=%D0%A0%D0%B0%D0%B7%D1%80%D0%B0%D0%B1%D0%BE%D1%82%D1%87%D0%B8%D0%BA+python&salary=&currency_code=RUR&experience=doesNotMatter&order_by=relevance&search_period=0&items_on_page=50'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'}
r = req.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'lxml')
total_pages = int(soup.findAll(class_='bloko-button')[-4].text)

for p in range(0, total_pages):
    url = f'https://hh.ru/search/vacancy?no_magic=true&L_save_area=true&text=%D0%A0%D0%B0%D0%B7%D1%80%D0%B0%D0%B1%D0%BE%D1%82%D1%87%D0%B8%D0%BA+python&salary=&currency_code=RUR&experience=doesNotMatter&order_by=relevance&search_period=0&items_on_page=50?page={p}'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'}
    r = req.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'lxml')
    
    vac = soup.findAll('a', class_='serp-item__title')
    
    for v in vac:
        link = v.get('href').split('?')[0]
        vacancies.append(link)
    
print('get all vacancies')

data = {'data':[]}

# парсим нужные данные
for i in vacancies:
    url = i
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'}
    r = req.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'lxml')
    
    try:
        title = soup.find('div', class_='vacancy-title').find('h1', class_='bloko-header-section-1').text
    except:
        pass
    
    try:
        exp = soup.find(attrs={'data-qa':'vacancy-experience'}).text
    except:
        pass
    
    try:
        salary = str(soup.find(attrs={'data-qa':'vacancy-salary-compensation-type-net'}).text).replace(u'\xa0', ' ')
    except:
        pass
    
    try:
        region = soup.find(attrs={'data-qa':'vacancy-view-raw-address'}).text.split(',')[0]
    except:
        pass
    
    data['data'].append({'title':title,
                        'work experience': exp,
                        'salary': salary, 
                        'region': region})
    
with open("data.json", "w") as file:
    json.dump(data, file, ensure_ascii=False)
```

```python
# 3. Перевод арабского числа в римское

def roman_number(arabic_number):
    dd = {1:'I',
    5:'V',
    10:'X',
    50:'L',
    100:'C',
    500: 'D',
    1000:'M'}
    
    
    num = arabic_number

    if len(str(num)) == 1:
        a = num
    elif len(str(num)) == 2:
        a = int(str(num)[-1])
        b = int(str(num)[-2])*1
    elif len(str(num)) == 3:
        a = int(str(num)[-1])
        b = int(str(num)[-2])*10
        c = int(str(num)[-3])*10
    elif len(str(num)) == 4:
        a = int(str(num)[-1])
        b = int(str(num)[-2])*10
        c = int(str(num)[-3])*100
        d = int(str(num)[-4])*1000
        
        
    if a in [1,5]:
        e = dd[a]
    elif a in [2,3]:
        e = dd[1]*a
    elif a in [6, 7, 8]:
        e = dd[5]+(a-5)*dd[1]
    elif a == 4:
        e = dd[1]+dd[5]
    else:
        e = dd[1]+dd[10]


    # десятки
    if b in [10,50]:
        f = dd[b]
    elif b in [20,30]:
        f = dd[10]*int(str(b)[0])
    elif b in [60, 70, 80]:
        f = dd[50]+(int(str(b-50)[0]))*dd[10]
    elif b == 40:
        f = dd[10]+dd[50]
    else:
        f = dd[10]+dd[100]

    # сотни
    if c in [100,500]:
        g = dd[c]
    elif c in [200,300]:
        g = dd[100]*int(str(c)[0])
    elif c in [600, 700, 800]:
        g = dd[500]+(int(str(c-500)[0]))*dd[100]
    elif c == 400:
        g = dd[100]+dd[500]
    else:
        g = dd[100]+dd[1000]

    #тысячи
    if d == 1000:
        h = dd[d]
    elif d == 2000:
        h = dd[1000]*2


    if len(str(num))==1:
        arabic_number = e
    elif len(str(num))==2:
        arabic_number = f+e
    elif len(str(num))==3:
        arabic_number = g+f+e
    elif len(str(num))==4:
        arabic_number = h+g+f+e


    return arabic_number
```

```python
# 4. Валидность скобок
def task(string):
    d = {'[':0, ']':0, '(':0, ')':0, '{':0, '}':0}
    for i in string:
        d[i] = d[i]+1
    
    if d['['] == d[']'] and d['('] == d[')'] and d['{'] == d['}']:
        print(True)
    else:
        print(False)
```

```python
# 5. Умножить два бинарных числа в формате строк

def bools_addition(X1, X2):
#     Y1
    l = []
    for i in str(X1):
        l.append(int(i))
    Y1 = 0
    index = len(l)-1
    for i in l:
        Y1 += i*(2**index)
        index = index-1    
# Y2
    l = []
    for i in str(X2):
        l.append(int(i))
    Y2 = 0
    index = len(l)-1
    for i in l:
        Y2 += i*(2**index)
        index = index-1
        
# X
    Y = Y1*Y2
    X = ''
 
    while Y > 0:
        X = str(Y % 2) + X
        Y = Y // 2
        
    return X
    
```
